```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv(r'Mall_Customers.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data= data.iloc[:,3:].values
data
```

```python
wcss = []
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init= 'k-means++', random_state=10)
y_means = kmeans.fit_predict(data)
```

```python
y_means
```

```python
data[y_means==0,1]
```

```python
plt.scatter(data[y_means==0,0], data[y_means==0,1],s=100,color='red',label='cluster 1')
plt.scatter(data[y_means==1,0], data[y_means==1,1],s=100,color='blue',label='cluster 2')
plt.scatter(data[y_means==2,0], data[y_means==2,1],s=100,color='green',label='cluster 3')
plt.scatter(data[y_means==3,0], data[y_means==3,1],s=100,color='yellow',label='cluster 4')
plt.scatter(data[y_means==4,0], data[y_means==4,1],s=100,color='orange',label='cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='black', label='Centroids')
plt.title('Cluster of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
