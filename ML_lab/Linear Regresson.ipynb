# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
```

```python
data=pd.read_csv(r'Salary_Data.csv')
data
```

```python
X=data.iloc[:,:-1].values
Y=data.iloc[:,-1:].values
#X,Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=  train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model  import LinearRegression as lr
regressor=lr()
```

```python
regressor.fit(X_train,Y_train)
```

```python
pred=regressor.predict(X_test)
```

```python
mpl.scatter(X_train,Y_train,color='red')
mpl.plot(X_train,regressor.predict(X_train),color='blue')
mpl.title('Salary vs Experience(training set)')
mpl.xlabel('Years of experience')
mpl.ylabel('Salary')
mpl.show()
```

```python
mpl.scatter(X_test,Y_test,color='red')
mpl.plot(X_test,regressor.predict(X_test),color='blue')
mpl.title('Salary vs Experience(training set)')
mpl.xlabel('Years of experience')
mpl.ylabel('Salary')
mpl.show()
```
