## Without inbuilt library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("dataset.csv")
print(data)
```

```python
X = data.iloc[:,:-1].values 
y = data.iloc[:,-1:].values

plt.scatter(X, y)
plt.title('Scatter Plot ')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
w,b = 0.1, 0.1  # Initial weight
learning_rate = 0.001
epochs = 1000
```

```python
for i in range(epochs):
    y_pred = w * X_train + b
    loss = 0.5*(np.mean(np.square(y_train-y_pred)))
    dw = np.mean(X_train*(-y_train+y_pred))
    db = np.mean(-y_train+y_pred)
    w = w - learning_rate*dw
    b = b - learning_rate*db
    if i%100 == 0 :
        print('Loss',loss)
```

```python
print(f'Final values - w: {w}, b: {b}')
```

```python
y_pred= w*X_test + b
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red')  # Plotting the regression line
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Fitted Line')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```
