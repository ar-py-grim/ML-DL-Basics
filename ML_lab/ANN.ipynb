```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = pd.read_csv(r"Churn_Modelling.csv") 
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop(columns= ['RowNumber', 'CustomerId', 'Surname'], axis =1 , inplace = True) 
data
```

```python
data['Geography']= data['Geography'].map({'France' : 0, 'Germany' : 1, 'Spain' : 2}) 
data['Gender']= data['Gender'].map({'Male' : 0, 'Female' : 1})
```

```python
data
```

```python
# Features

x = data.iloc[:, :-1]
x = np.array(x)
x
```

```python
# Output

y=data.iloc[:, -1]
y=np.array(y)
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state=0) 
```

```python
sc=StandardScaler() 
x_train= sc.fit_transform(x_train) 
x_test = sc.transform(x_test) 
```

```python
# Creating model 

ann = tf.keras.models.Sequential() 
# Adding 2 hidden layers 
ann.add(tf.keras.layers.Dense(units=6, activation='relu')) 
ann.add(tf.keras.layers.Dense(units=6, activation='relu')) 
# Adding final layer 
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid')) 
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) 
ann_hist = ann.fit(x_train, y_train, validation_split= 0.2, validation_data= (x_test, y_test), epochs = 100)
```

```python
y_pred = ann.predict(x_test)
y_pred
```

```python
y_pred = (y_pred>0.5) 
y_pred
```

```python
cm= confusion_matrix(y_test, y_pred) 
print(cm)
accuracy_score(y_test, y_pred)
```
