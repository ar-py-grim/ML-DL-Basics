```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
from sklearn.model_selection import train_test_split
from sklearn.linear_model  import LogisticRegression as logr
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
data=pd.read_csv(r'framingham.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.fillna(data.mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
attribute = data.iloc[:,:-1]
attribute = np.array(attribute)
#attribute
```

```python
target = data.iloc[:,-1:]
target = np.array(target)
#target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(attribute,target,test_size=0.25,random_state=1)
```

```python
log_reg = logr()
```

```python
log_reg.fit(X_train,Y_train)
```

```python
pred=log_reg.predict(X_test)
```

```python
accuracy_score(Y_test,pred)
```

# Random Forest v Logistic Regression

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=2) 
classifier.fit(X_train,Y_train)
```

```python
y_pred_rf=classifier.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred_rf)
```
