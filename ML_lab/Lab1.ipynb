```python
import numpy as np

# to see arguments of function go inside common brackets of func. and press Shift+Tab
```

```python
a1 = np.array([1,2,3,6])
print(f" 1D array: {a1}, len: {len(a1)}, dimension: {a1.ndim}, type: {a1.dtype}, shape: {a1.shape}")
```

```python
a2 = np.array([[1.0, 2.7, 6.8], [8.5, 9.5, 55]])
print(f" 2D array: {a2}, len: {len(a2)}, dimension: {a2.ndim}, type: {a2.dtype}, shape: {a2.shape}")
```

```python
a3 = np.array([[['1', '2.7', '6.8'], ['8.5', '9.5', '55']],[['3','4','1'],['3','67','54']]])
print(f" 3D array: {a3}, len: {len(a3)}, dimension: {a3.ndim}, type: {a3.dtype}, shape: {a3.shape} \n")

print(f"max no. in a2 array is: {np.max(a2)}")
```

```python
print(f"sum of elements in a2 is: {np.sum(a2)}")
print(f"sum of elements in a2 column is: {np.sum(a2, axis=0)}")
print(f"sum of elements in a2 row is: {np.sum(a2, axis=1)}")
```

```python
a6 = np.array([[1,2,3,33,44],[4,5,6,11,22],[7,8,9,4,21],[56,34,10,11,12],[560,340,100,11,120],[506,304,100,101,102]])
print(a6)
```

```python
print(f"last 4 no. of last row is: {a6[5:,1:]}")
```

```python
# to reshape a numpy array
# total elements in a2 is 6

a7 = a2.reshape(6,1)
print(f" reshaped column array is: {a7} \n")

a8 = a2.reshape(1,6)
print(f" reshaped row array is: {a8} \n")

a9 = a2.reshape(2,3,1)
print(f"reshaped 3D array is {a9}")
```

```python
# linspace
# 50 is total numbers to generate

a11 = np.linspace(1,10,50, retstep=True)
print(a11)

a13 = np.linspace(start=[1,2],stop=[9,10],num=9)
print(f"generate 2D array: {a13}")
```

```python
# np.arrange function
# here 3 represents difference between consecutuve terms

a12 = np.arange(1,100,3)
a12
```
